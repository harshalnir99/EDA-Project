```python
# Plot Heatmap graph to check for correlation

sns.heatmap(df.corr(),annot=True,cmap='icefire',linewidths=0.3)
fig=plt.gcf()
fig.set_size_inches(10,10)
plt.title("corr between variable",color='black',size=15)
plt.show()
```

```python
df.columns
```

```python
# To check crime rate in 2016

plt.figure(figsize=(8,8))
sns.barplot(x='Category', y ='2016', data=df)
plt.title("Crime Rate in 2016")
plt.grid(axis='y')
plt.show()
```

```python
# Plot Graph State/UT wise vs crime rate in year 2016

plt.figure(figsize=(15,8))
sns.barplot(x='State/UT', y ='2016', data=df )
plt.title("State/UT wise Crime Rate in 2016")
plt.grid(axis='y')
plt.xticks(rotation= 90)
plt.show()
```

```python
# To check crime rate in 2017

plt.figure(figsize=(8,8))
sns.barplot(x='Category', y ='2017', data=df )
plt.title("Crime Rate in 2017")
plt.grid(axis='y')
plt.show()
```

```python
# Plot Graph State/UT wise vs crime rate in year 2017

plt.figure(figsize=(15,8))
sns.barplot(x='State/UT', y ='2017', data=df )
plt.title("State/UT wise Crime Rate in 2017")
plt.grid(axis='y')
plt.xticks(rotation= 90)
plt.show()
```

```python
# To check crime rate in 2018

plt.figure(figsize=(8,8))
sns.barplot(x='Category', y ='2018', data=df )
plt.title("Crime Rate in 2018")
plt.grid(axis='y')
plt.show()
```

```python
# Plot Graph State/UT wise vs crime rate in year 2018

plt.figure(figsize=(15,8))
sns.barplot(x='State/UT', y ='2018', data=df )
plt.title("State/UT wise Crime Rate in 2018")
plt.grid(axis='y')
plt.xticks(rotation= 90)
plt.show()
```

```python
df.rename(columns={"Percentage Share of State/UT (2018)":"Percentage Share" ,
           "Mid-Year Projected Population (in Lakhs) (2018)+":"Projected Population",
          "Rate of Total Cyber Crimes (2018)++" :"rate of total Cyber crimes"} , inplace=True)
```

```python
df['Total cyber crime'] = df["2016"]+df["2017"]+df["2018"]
```
